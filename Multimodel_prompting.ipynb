```python
## MULTIMODAL SALES CHART ANALYSIS WITH STATISTICS

from _pipeline import create_payload, model_req
import base64

#### (1) Load Sales Chart Image
IMAGE_PATH = r"C:\Users\patel\Downloads\1.png"
with open(IMAGE_PATH, "rb") as image_file:
    encoded_image = base64.b64encode(image_file.read()).decode("utf-8")

#### (2) Updated Prompt for Statistical + Business Analysis
PROMPT = f"""
Explain this image.
"""

#### (3) Configure the Model Request
payload = create_payload(
    target="ollama",
    model="llava:7b",  # Ensure model supports images
    prompt=PROMPT, 
    image=encoded_image,
    temperature=0.7, 
    num_ctx=250, 
    num_predict=400
)

### Send to the model
time, response = model_req(payload=payload)
print(response)
if time:
    print(f'Time taken: {time}s')
```
